```python
def preprocessing(df, column_name):
    df[column_name] = df[column_name].str.lower()
    df[column_name] = df[column_name].str.replace("'ll", " will")
    df[column_name] = df[column_name].str.replace("'re", " are")
    df[column_name] = df[column_name].str.replace("’re", " are")
    df[column_name] = df[column_name].str.replace(".", " .")
    df[column_name] = df[column_name].str.replace(",", " ,")
    df[column_name] = df[column_name].str.replace("?", " ?")
    df[column_name] = df[column_name].str.replace("!", " !")
    df[column_name] = df[column_name].str.replace("'", "")
    return df[column_name]
```

```python
def candidateHistory_test(scenario_df):
    utterances = []
    
    scenarioNum = list(scenario_df['scenario'].values)[0]
    
    
    history_values = scenario_df[scenario_df['scenario']==scenarioNum]['en'].values

    # 1번 context // 1개 history
    contextNum =0
    # eat 영역에서 19 랜덤 추출 + 선택된 시나리오ID에 다음 대화 추가(정답 LABEL)
    candidate = list(np.append(np.random.choice(scenario_df['en'], 19), history_values[contextNum+1]))
    
    history_temp = []
    history_temp.append(scenario_df[scenario_df['scenario']==scenarioNum]['en'].values[(contextNum)])

    utterance = {"candidates": candidate, "history":history_temp}
    utterances.append(utterance)

    # 2번 context // 3개 history
    contextNum =3
    candidate = list(np.append(np.random.choice(scenario_df['en'], 19), history_values[contextNum]))
    history = list(scenario_df[scenario_df['scenario']==scenarioNum]['en'].values[: contextNum])

    utterance = {"candidates": candidate, "history":history}
    utterances.append(utterance)


    # 3번 context // 4개 history -> 홀수 개의 history가 아니므로 가짜 history 1개 추가(정답 sentence없음) 
    # 임의의 가짜 history를 추가 하여 5개의 history 맞춤.
    candidate = list(np.random.choice(scenario_df['en'], 20)) #임의의 candidate
    history = list(scenario_df[scenario_df['scenario']==scenarioNum]['en'].values) 
    history.append(scenario_df[scenario_df['scenario']==scenarioNum]['en'].values[0])    # 임의의 가짜 히스토리    

    utterance = {"candidates": candidate, "history":history}
    utterances.append(utterance)
    
    return utterances
```

## Encode Labels
Encode target labels that is in 'minor' attribute

```python
# cccc
minor_list = np.unique(df['minor'].values)

def merge_context(sap_df):
    merge = []
    for minor in minor_list:
            minor_sap_df = sap_df[sap_df['minor']==minor]['scenario']
            uniq_scenario_num = np.unique(minor_sap_df.values)

            for num in uniq_scenario_num:
                context = sap_df[(sap_df['scenario'] == num)& (sap_df['minor'] == minor)]['en'].values
                
                context = " ".join(context)
                merge_dict = {'minor':minor, 'scenario': num, 'en': context}
                merge.append(merge_dict)
                
    return merge
```
